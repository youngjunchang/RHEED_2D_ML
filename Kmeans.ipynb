```python
m = np.zeros((20,fshape[0],fshape[1]))
k=0
for i in np.arange(5):
    for j in np.arange(i+2):
        m[k] = kmean_image(index[i][j],rvid_gf,rvid_gray,save=0).reshape(fshape[0],fshape[1])
        k = k+1

kmin = np.amin(m)
kmax = np.amax(m)
```

```python
fig = plt.figure(figsize=(15,10),dpi=200)
k = 0

for i in np.arange(5):
    for j in np.arange(i+2):
        plt.subplot(5, 6, i*6+j+1),plt.imshow(m[k],vmin=0,vmax=kmax,cmap="rainbow"),plt.axis("off")
        k = k+1
        
plt.subplots_adjust(wspace=0.1, hspace=0)
plt.show()
#plt.savefig("07_"+name+"_kmeans_image_01.png")
```
